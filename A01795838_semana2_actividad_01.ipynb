##Funcion para mostrar los resultados en formato ameno ‚õÖ

```python
#importamos 2 librerais mas solo para fines esteticos, agenos al analisis de regex
import html  # Para escapar caracteres especiales en los elementos
from IPython.display import display, HTML

def imprimir_tabla_resultados(elementos, num_columnas=3):
    """
    Muestra una lista de elementos ordenados en formato de tabla HTML,
    optimizado para visualizaci√≥n en notebooks (incluyendo GitHub).

    Args:
        elementos: Lista de elementos a mostrar (se convertir√°n a string).
        num_columnas: N√∫mero de columnas para la tabla.

    Returns:
        int: Cantidad total de elementos mostrados.
    """
    # Asegurar que todos los elementos sean strings y ordenarlos
    elementos_ordenados = sorted([str(elem) for elem in elementos])
    total_elementos = len(elementos_ordenados)

    if total_elementos == 0:
        # Si no hay elementos, mostrar un mensaje simple (HTML)
        display(HTML("<p>No hay elementos para mostrar.</p>"))
        return 0

    # --- Construcci√≥n de la cadena HTML ---

    # Estilos CSS b√°sicos para la tabla (puedes personalizarlos)
    # Usamos un estilo simple y limpio compatible con el modo claro/oscuro de GitHub
    css_styles = """
    <style>
        .github-table {
            border-collapse: collapse;
            margin: 1em 0;
            font-size: 0.9em;
            font-family: -apple-system, BlinkMacSystemFont, "Segoe UI", Helvetica, Arial, sans-serif, "Apple Color Emoji", "Segoe UI Emoji";
            width: auto; /* Opcional: '100%' para ancho completo */
            box-shadow: 0 0 5px rgba(0, 0, 0, 0.1); /* Sombra sutil */
        }
        .github-table th, .github-table td {
            border: 1px solid #dfe2e5; /* Borde gris claro, como GitHub */
            padding: 8px 12px;
        }
        .github-table td.col-centered {
            text-align: center;
        }
        .github-table td.col-left {
            text-align: left;
        }
        /* Estilo para el mensaje de total */
        .total-count-message {
            margin-top: 1em;
            font-family: -apple-system, BlinkMacSystemFont, "Segoe UI", Helvetica, Arial, sans-serif, "Apple Color Emoji", "Segoe UI Emoji";
        }
        .total-count-highlight {
            color: #E67E22; /* Un tono naranja similar al ANSI 208 */
            font-weight: bold;
        }
    </style>
    """

    # Inicio de la tabla HTML
    html_string = css_styles + '<table class="github-table">\n<tbody>\n'

    # Determinar la clase CSS para la alineaci√≥n de celdas
    cell_class = "col-left" if num_columnas == 1 else "col-centered"

    # Iterar sobre los elementos para construir las filas de la tabla
    for i in range(0, total_elementos, num_columnas):
        html_string += "  <tr>\n"
        elementos_fila = elementos_ordenados[i:i + num_columnas]

        # A√±adir cada elemento como una celda <td>
        for elem in elementos_fila:
            # Escapar caracteres HTML como '<', '>' para evitar problemas
            safe_elem = html.escape(elem)
            html_string += f'    <td class="{cell_class}">{safe_elem}</td>\n'

        # Si la √∫ltima fila no est√° completa, a√±adir celdas vac√≠as <td>
        celdas_faltantes = num_columnas - len(elementos_fila)
        if celdas_faltantes > 0:
            html_string += f'    <td class="{cell_class}"></td>\n' * celdas_faltantes

        html_string += "  </tr>\n"

    # Cierre de la tabla
    html_string += "</tbody>\n</table>\n"

    # A√±adir el mensaje del total (fuera de la tabla, como un p√°rrafo)
    # Se aplica el estilo con un <span>
    total_message = (
        f'<p class="total-count-message">El n√∫mero total de resultados obtenidos es: '
        f'<span class="total-count-highlight">{total_elementos}</span></p>'
    )
    html_string += total_message

    # --- Mostrar el HTML generado en el notebook ---
    display(HTML(html_string))

    # Retornar la cantidad total de elementos
    return total_elementos
```

*   **Pregunta 1.**

Busca y elimina todos los saltos de l√≠nea '\n' que se encuentran al final de cada comentario.

Una vez finalizado, imprime los primeros 10 comentarios del resultado obtenido.


```python
# Inicializamos una lista para almacenar los resultados
lista_resultados = []

# Procesamos los documentos para eliminar saltos de l√≠nea al final de cada registro
textos_procesados = [re.sub(r'\n$', '', registro) for registro in docs]  # Elimina los saltos de l√≠nea al final de cada string

# Mostramos los primeros 10 registros en formato de tabla usando nuestra funci√≥n mejorada
total_mostrados = imprimir_tabla_resultados(textos_procesados[:10], num_columnas=1)
```

```python
# Imprimimos un mensaje adicional sobre los datos mostrados en color azul
print(f"\033[34müìù Se han mostrado {total_mostrados} de {len(textos_procesados)} registros totales.\033[0m")
```

*   **Pregunta 2.**  

Busca e imprime todas las palabras que terminan con dos o m√°s signos de admiraci√≥n seguidos, por ejemplo "!!!".

Debes imprimir tanto la palabra como la totalidad de signos de admiraci√≥n que le siguen.

Indica cu√°ntos resultados obtuviste.



```python
# Inicializamos una lista para almacenar palabras que contienen m√∫ltiples exclamaciones
palabras_exclamativas = []

# Iteramos por cada texto procesado para buscar patrones espec√≠ficos
for texto in textos_procesados:
    # Buscamos palabras seguidas de dos o m√°s signos de exclamaci√≥n y las a√±adimos a la lista
    palabras_exclamativas.extend(re.findall(r'\b\w+!{2,}', texto))

# Mostramos las palabras encontradas en una tabla de 5 columnas
total_palabras_exclamativas = imprimir_tabla_resultados(palabras_exclamativas, num_columnas=5)

# Guardamos el resultado en nuestra lista de resultados para an√°lisis posterior
lista_resultados.append(("Palabras con dos o m√°s exclamaciones", total_palabras_exclamativas))
```

```python
# Imprimimos un mensaje adicional sobre los datos mostrados en color azul
print(f"\033[34müìù Se encontraron {total_palabras_exclamativas} palabras con m√∫ltiples exclamaciones en el corpus analizado.\033[0m")
```

*   **Pregunta 3.**  

Busca e imprime todas las palabras que est√°n escritas totalmente en may√∫sculas. Cada coincidencia debe ser una sola palabra.

Indica cu√°ntas palabras encontraste.



```python
# Inicializamos un conjunto para almacenar palabras en may√∫sculas (sin duplicados)
palabras_mayusculas = set()

# Iteramos por cada texto procesado para buscar palabras en may√∫sculas
for texto in textos_procesados:
    # Actualizamos el conjunto con palabras formadas solo por letras may√∫sculas
    palabras_mayusculas.update(re.findall(r'\b[A-Z]+\b', texto))

# Mostramos las palabras encontradas en una tabla de 8 columnas
total_palabras_mayusculas = imprimir_tabla_resultados(palabras_mayusculas, num_columnas=8)

# Guardamos el resultado en nuestra lista de resultados para an√°lisis posterior
lista_resultados.append(("Palabras en may√∫sculas", total_palabras_mayusculas))
```

```python
 # Imprimimos un mensaje adicional para contextualizar en color azul
print(f"\033[34müî† Se encontraron {total_palabras_mayusculas} palabras escritas completamente en may√∫sculas.\033[0m")
```

*   **Pregunta 4.**  

Busca e imprime los comentarios en donde todos los caracteres alfab√©ticos (letras) est√°n en may√∫sculas.

Cada coincidencia encontrada debe ser todo el comentario/enunciado.

Indica cu√°ntos resultados obtuviste.


```python
# Inicializamos una lista para almacenar comentarios escritos completamente en may√∫sculas
textos_completamente_mayusculas = []

# Iteramos por cada texto procesado para encontrar los que est√°n completamente en may√∫sculas
for texto in textos_procesados:
    # Verificamos si el texto completo coincide con el patr√≥n de solo may√∫sculas, espacios y s√≠mbolos
    if re.fullmatch(r'[A-Z\s\W]+', texto):
        textos_completamente_mayusculas.append(texto)

# Mostramos los comentarios encontrados en una tabla de 1 columna
total_textos_mayusculas = imprimir_tabla_resultados(textos_completamente_mayusculas, num_columnas=1)

# Guardamos el resultado en nuestra lista de resultados para an√°lisis posterior
lista_resultados.append(("Comentarios en may√∫sculas", total_textos_mayusculas))
```

#**Parte 3. Proceso de limpieza.**

*   **Pregunta 13.**  

Ahora realiza un proceso de limpieza del corpus que incluya los siguientes procesos:

*   Solo se deben considerar caracteres alfab√©ticos. Es decir, se eliminan todos los signos de puntuaci√≥n y caracteres especiales.
*   Todos los caracteres alfab√©ticos se transforman a min√∫sculas.
*   Se deben eliminar todos los espacios en blanco adicionales que se puedan encontrar en cada comentario.

Al finalizar dicho proceso de limpieza, imprime el resultado de los primeros 10 comentarios resultantes.
   




```python
def normalizar_textos(textos_originales):
    """
    Funci√≥n que limpia y normaliza una lista de textos.

    Args:
        textos_originales: Lista de textos a normalizar

    Returns:
        Lista de textos normalizados y limpios
    """
    # Inicializamos una lista para almacenar los textos normalizados
    textos_normalizados = []

    # Procesamos cada texto de la lista original
    for texto in textos_originales:
        # Eliminamos todos los caracteres que no sean letras o espacios
        texto_solo_letras = re.sub(r'[^a-zA-Z\s]', '', texto)  # Solo mantener caracteres alfab√©ticos (y espacios)

        # Convertimos todo el texto a min√∫sculas para estandarizaci√≥n
        texto_minusculas = texto_solo_letras.lower()  # Convertir a min√∫sculas

        # Eliminamos espacios m√∫ltiples y normalizamos a un solo espacio
        texto_normalizado = re.sub(r'\s+', ' ', texto_minusculas)  # Eliminar espacios en blanco adicionales

        # A√±adimos el texto procesado a la lista de resultados
        textos_normalizados.append(texto_normalizado)

    # Imprimimos un mensaje informativo sobre el proceso realizado
    print(f"\033[34müßπ Se han normalizado \033[38;5;208m\033[1m{len(textos_normalizados)}\033[0m\033[34m textos, eliminando caracteres especiales y estandarizando a min√∫sculas.\033[0m")

    # Devolvemos la lista de textos ya procesados
    return textos_normalizados
```

```python
# Aplicamos el proceso de normalizaci√≥n a todos los textos
textos_normalizados = normalizar_textos(textos_procesados)

# Mostramos una vista previa de los primeros 10 textos normalizados
total_textos_mostrados = imprimir_tabla_resultados(textos_normalizados[:10], num_columnas=1)
# Imprimimos un mensaje adicional para explicar la visualizaci√≥n
print(f"\033[34müëÅÔ∏è Se muestran los primeros {total_textos_mostrados} textos despu√©s de su normalizaci√≥n.\033[0m")
```

*   **Pregunta 14.**  

Con el resultado de la limpieza obtenido en la pregunta anterior, realiza ahora un proceso de tokenizaci√≥n por palabras del corpus.

Es decir, al final de este proceso de tokenizaci√≥n, debes tener como resultado una lista de listas, donde cada comentario estar√° tokenizado por palabras.

Al terminar calcula el total de tokens obtenido en todo el corpus.

```python
# Tokenizamos cada texto dividi√©ndolo en palabras individuales
textos_tokenizados = [texto.split() for texto in textos_normalizados]

# Calculamos el n√∫mero total de tokens (palabras) en todos los textos
conteo_total_tokens = sum(len(tokens) for tokens in textos_tokenizados)

# Guardamos el resultado en nuestra lista para an√°lisis posterior
lista_resultados.append(("Total de tokens", conteo_total_tokens))

# Imprimimos un mensaje con el n√∫mero total de tokens resaltado en naranja
print(f"üî§ Total de tokens encontrados: \033[38;5;208m\033[1m{conteo_total_tokens}\033[0m")
```

*   **Pregunta 15.**  

Finalmente, en este ejercicio definiremos nuestro conjunto de palabras "stopwords", las cuales deber√°s eliminar de todo el corpus.

Recuerda que ejemplos de stopwords son art√≠culos, adverbios, conectivos, etc√©tera, que tienen frecuencias de aparici√≥n muy altas en cualquier documento, pero que no brindan mucho significado en cuanto al significado de un enunciado.

Con base a la lista de stopwords que se te proporciona, realiza un proceso de limpieza eliminando todas estas palabras del corpus obtenido en el ejercicio anterior.

Obtener cu√°ntos tokens/palabras quedan finalmente en todo el corpus.

Obtener cu√°ntos de estos tokens/palabras son diferentes, es decir, cu√°ntos tokens √∫nicos tendr√° lo que llamaremos m√°s adelante nuestro vocabulario.

```python
# Definimos una lista de palabras vac√≠as (stopwords) en ingl√©s para eliminar del an√°lisis
# Estas son palabras muy comunes que generalmente no aportan significado relevante
mis_stopwords = [
    # Pronombres personales
    'i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you', 'your', 'yours',
    'he', 'him', 'his', 'himself', 'she', 'her', 'hers', 'herself', 'it', 'its', 'itself',
    'they', 'them', 'their', 'theirs', 'themselves',

    # Pronombres interrogativos/relativos
    'what', 'which', 'who', 'whom',

    # Pronombres demostrativos
    'this', 'that', 'these', 'those',

    # Verbos auxiliares y comunes
    'am', 'is', 'are', 'was', 'were', 'be', 'been', 'being', 'have', 'has', 'had',
    'having', 'do', 'does', 'did', 'doing',

    # Art√≠culos y preposiciones
    'a', 'an', 'the', 'and', 'but', 'if', 'or', 'because', 'as', 'until', 'while',
    'of', 'at', 'by', 'for', 'with', 'about', 'against', 'between', 'into', 'through',
    'during', 'before', 'after', 'above', 'below', 'to', 'from', 'up', 'down', 'in',
    'out', 'on', 'over', 'under',

    # Adverbios y otros
    'again', 'further', 'then', 'once', 'here', 'there', 'when', 'where', 'why', 'how',
    'all', 'any', 'both', 'each', 'few', 'more', 'most', 'other', 'some', 'such', 'only',
    'own', 'same', 'so', 'than', 'too', 'very',

    # Contracciones y otros
    's', 't', 'can', 'will', 'just', 'should', 'now', 'll'
]

# Imprimimos informaci√≥n sobre la lista de palabras vac√≠as
print(f"üö´ Se han cargado \033[38;5;208m\033[1m{len(mis_stopwords)}\033[0m palabras vac√≠as en ingl√©s que ser√°n filtradas del an√°lisis.")
```

```python
# Filtramos las palabras vac√≠as (stopwords) de nuestros textos tokenizados
textos_filtrados = []
for tokens_texto in textos_tokenizados:
    # Para cada texto, conservamos solo las palabras que no est√°n en nuestra lista de palabras vac√≠as
    textos_filtrados.append([palabra for palabra in tokens_texto if palabra not in mis_stopwords])

# Calculamos el n√∫mero total de tokens que quedaron despu√©s de filtrar palabras vac√≠as
total_tokens_filtrados = sum(len(tokens) for tokens in textos_filtrados)

# Guardamos el resultado en nuestra lista para an√°lisis posterior
lista_resultados.append(("Total de tokens restantes", total_tokens_filtrados))

# Imprimimos un mensaje con el n√∫mero de tokens filtrados resaltado en naranja
print(f"üìä Total de tokens despu√©s de filtrar palabras vac√≠as: \033[38;5;208m\033[1m{total_tokens_filtrados}\033[0m")

# Calculamos el vocabulario √∫nico (conjunto de palabras diferentes) despu√©s de filtrar
vocabulario_unico = set([palabra for texto in textos_filtrados for palabra in texto])
total_palabras_unicas = len(vocabulario_unico)

# Guardamos el resultado en nuestra lista para an√°lisis posterior
lista_resultados.append(("Total de tokens √∫nicos", total_palabras_unicas))

# Imprimimos un mensaje con el tama√±o del vocabulario resaltado en naranja
print(f"üîç Total de palabras √∫nicas (vocabulario): \033[38;5;208m\033[1m{total_palabras_unicas}\033[0m")

# Calculamos la riqueza l√©xica (proporci√≥n de palabras √∫nicas respecto al total)
if total_tokens_filtrados > 0:
    riqueza_lexica = total_palabras_unicas / total_tokens_filtrados * 100
    print(f"üìö Riqueza l√©xica del corpus: \033[38;5;208m\033[1m{riqueza_lexica:.2f}%\033[0m")
```

```python
# Imprimimos un encabezado para el resumen de resultados
print("\n" + "="*60)
print("üìã \033[1mRESUMEN DE RESULTADOS OBTENIDOS\033[0m")
print("="*60)

# Calculamos el ancho m√°ximo necesario para la primera columna (descripci√≥n)
ancho_descripcion = max([len(str(descripcion)) for descripcion, _ in lista_resultados], default=30) + 2

# Formateamos cada fila de resultados
for descripcion, valor in lista_resultados:
    # Creamos un formato alineado a la izquierda para la descripci√≥n
    descripcion_formateada = f"{descripcion:{ancho_descripcion}}"

    # Imprimimos cada resultado con el valor resaltado en naranja
    print(f"‚îÇ {descripcion_formateada} ‚îÇ \033[38;5;208m\033[1m{valor:,}\033[0m")

    # A√±adimos un separador de l√≠nea despu√©s de cada resultado
    print(f"‚îú{'‚îÄ' * ancho_descripcion}‚îº‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚î§")

# Calculamos algunos estad√≠sticos adicionales si es posible
if len(lista_resultados) > 0:
    # Calculamos el promedio de los valores num√©ricos
    valores = [valor for _, valor in lista_resultados if isinstance(valor, (int, float))]
    if valores:
        promedio = sum(valores) / len(valores)
        print(f"\nüìà Promedio de los valores num√©ricos: \033[38;5;208m\033[1m{promedio:,.2f}\033[0m")

# Imprimimos un mensaje de conclusi√≥n
print("\nüèÅ An√°lisis de texto completado con √©xito.")
```

En ese contexto, las **expresiones regulares (regex)** funcionan como un bistur√≠ que a√≠sla y normaliza patrones, mientras la depuraci√≥n sistem√°tica evita el temido efecto *garbage-in ‚Üí garbage-out*.

- **Tokenizaci√≥n enriquecida con regex**: al detectar emojis, fechas o URLs mediante patrones, los vectores capturan mejor la sem√°ntica original.  
- **An√°lisis exploratorio guiado por patrones**: muestrear rese√±as con puntuaciones extremas o enlaces externos agiliza los *insights* preliminares.  
- **Impacto directo en la precisi√≥n**: estudios muestran ca√≠das de hasta un 20 % en *F1-score* cuando persisten duplicados y outliers.  
- **Escalabilidad frente al tsunami de datos**: cada minuto, los usuarios de internet a√±aden vol√∫menes descritos en *Data Never Sleeps 12.0* , demandando limpiezas *streaming* basadas en regex y deduplicaci√≥n.

## ‚öôÔ∏è ¬øPor qu√© las expresiones regulares marcan la diferencia?

1. **Detecci√≥n lineal**: motores como **RE2** garantizan complejidad temporal O(n), evitando bloqueos por retro-ciclos.  
2. **Extracci√≥n de entidades**: una sola l√≠nea basta para separar correos o n√∫meros de factura.  
3. **Normalizaci√≥n y validaci√≥n**: `re.sub` homologa formatos heterog√©neos de fechas o c√≥digos postales, reduciendo cardinalidad.  
4. **Optimizaci√≥n de recursos**: filtrar antes de cargar ahorra CPU y RAM cuando se procesan *logs* en tiempo real  .

## üßπ Limpieza de datos en bases masivas

| Reto | Riesgo sin depuraci√≥n | Estrategias con regex & co. |
|------|----------------------|-----------------------------|

| **Outliers** | M√©tricas distorsionadas | Clipping, winsorizaci√≥n, regex para unidades |
| **Inconsistencias de formato** | Fallos en *joins* y agregaciones  | `re.sub` para unificar fechas, monedas |
| **Velocidad de generaci√≥n (1,7 MB/s)** | ETL saturadas | Limpieza en *stream* con patrones compilados |
