```python
def print_table(headers, data):

    formatted_data = [[item] for item in data]
    print(tab.tabulate(formatted_data, headers=[headers], tablefmt="grid"))
```

ü§ì **Explicaci√≥n de expresi√≥n regular**

**`\n:`**Encuentra todos los saltos de l√≠nea.

`re.sub(pattern, repl, string)`: Reemplaza todas las ocurrencias de "pattern", por el reemplazo "repl", dentro de la cadena "string".

*   **Pregunta 12.**  

Busca e imprime todas las palabras que terminan en "ing" o "ed".

Indica la cantidad de palabras que encontraste de cada una.

ü§ì **Explicaci√≥n de expresi√≥n regular**

**`\b:`** Word boundary. Hace match con la posici√≥n entre un caracter alfanum√©rico (a-zA-Z0-9_) y uno no alfanum√©rico. En este caso, hace match con el inicio y el final de la palabra.

**`\w+:`** Hace match con una o m√°s ocurrencias de cualquier caracter alfanum√©rico (a-zA-Z0-9_).

**`ing:`** Hace match con los caracteres "ing".

**`ed:`** Hace match con los caracteres "ed".


`re.findall(pattern, string)`: Encuentra las ocurrencias que cumplen con el patr√≥n pattern, dentro de la cadena string.

**`Resultado`**: Recupera todas las palabras que terminan en ing o en ed.

```python
# Patr√≥n original
patron_ing = r"\b\w+ing\b"
patron_ed  = r"\b\w+ed\b"

palabras_ing = []
palabras_ed  = []

for comentario in comm_transf:
    # b√∫squeda sin distinguir may√∫sculas/min√∫sculas
    matches_ing = re.findall(patron_ing, comentario, re.IGNORECASE)
    matches_ed  = re.findall(patron_ed, comentario, re.IGNORECASE)
    # convertir todo a min√∫sculas para lista homog√©nea
    palabras_ing.extend([w.lower() for w in matches_ing])
    palabras_ed.extend([w.lower() for w in matches_ed])

# Mostrar resultados
total_ing = print_table('Palabras que terminan en ing', palabras_ing)
print(f'Cantidad de palabras que terminan en ing: {len(palabras_ing)}\n')

total_ed = print_table('Palabras que terminan en ed', palabras_ed)
print(f'Cantidad de palabras que terminan en ed: {len(palabras_ed)}')
```

*   Se deben eliminar todos los espacios en blanco adicionales que se puedan encontrar en cada comentario.

Al finalizar dicho proceso de limpieza, imprime el resultado de los primeros 10 comentarios resultantes.
   




```python
def limpieza(texto):
    texto = re.sub(r'[^a-zA-Z\s]', '', texto)
    texto = texto.lower()
    texto = re.sub(r'\s+', ' ', texto)
    return texto.strip()
```
